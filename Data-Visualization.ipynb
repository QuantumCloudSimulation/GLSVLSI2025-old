```python
import re


def extract_fidelity(filepath): 
    with open(filepath, 'r') as file:
        data = file.read()

    # Extract rows using regex
    pattern = r"device name: (\d+), .*?avg_fidelity: ([\d.]+) \+/- ([\d.]+)"
    matches = re.findall(pattern, data)

    # Parse and sort rows by device name (numeric order)
    sorted_data = sorted(matches, key=lambda x: int(x[0]))

    # Extract avg_fidelity and std_dev into separate lists
    avg_fidelity = [float(row[1]) for row in sorted_data]
    std_dev = [float(row[2]) for row in sorted_data]
    
    return avg_fidelity, std_dev
```

```python
# List of datasets (values) and corresponding standard deviations (std_dev)
avg_fidelity_1, std_dev_1 = extract_fidelity('results/counted-jobs-lambda-1.txt')
avg_fidelity_2, std_dev_2 = extract_fidelity('results/counted-jobs-lambda-2.txt')
avg_fidelity_3, std_dev_3 = extract_fidelity('results/counted-jobs-lambda-3.txt')
avg_fidelity_4, std_dev_4 = extract_fidelity('results/counted-jobs-lambda-4.txt')
avg_fidelity_5, std_dev_5 = extract_fidelity('results/counted-jobs-lambda-5.txt')
avg_fidelity_6, std_dev_6 = extract_fidelity('results/counted-jobs-lambda-6.txt')

# Combine datasets and std_dev into lists
datasets = [avg_fidelity_1, avg_fidelity_2, avg_fidelity_3, avg_fidelity_4, avg_fidelity_5, avg_fidelity_6]
std_devs = [std_dev_1, std_dev_2, std_dev_3, std_dev_4, std_dev_5, std_dev_6]

# Labels for each dataset
labels = ['Î» = 1', 'Î» = 2', 'Î» = 3', 'Î» = 4', 'Î» = 5', 'Î» = 6']

# Colorblind-friendly colors
colors = ["#E41A1C", "#377EB8", "#4DAF4A", "#FF7F00", "#984EA3", "#FFFF33"]

# Create x-axis positions for devices
num_devices = len(device_names)
x = np.arange(num_devices)

# Define the width of each bar
width = 0.15

plt.figure(figsize=(12, 3.5))

# Plot each dataset with error bars
for i, (dataset, std_dev) in enumerate(zip(datasets, std_devs)):
    plt.bar(x + i * width, dataset, width, label=labels[i], color=colors[i], yerr=std_dev, capsize=4)

# Add labels and title
plt.xlabel('QDevices', fontsize=14)
plt.ylabel('fidelity', fontsize=14)

# Set x-tick labels to device names
plt.xticks(ticks=x + width * 1.5, labels=device_names, rotation=45, ha='right', fontsize=14)

# Set y-axis limit for better visualization
plt.ylim(0, 1)

# Add legend to differentiate between datasets
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=12)

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the plot
# plt.savefig("fidelity-with-error-bars.png")

# Display the plot
plt.show()
```
